```python
# Tractar nuls
df_candidatures_partits['candidat_Dona_rel'] = np.nan
df_candidatures_partits['electe_Dona_rel'] = np.nan
df_candidatures_partits['candidat_Home_rel'] = np.nan
df_candidatures_partits['electe_Home_rel'] = np.nan

# Calcular on el divisor es > 0, establir condició
condicio_candidat = df_candidatures_partits['candidat_Total'] > 0
condicio_electe = df_candidatures_partits['electe_Total'] > 0

df_candidatures_partits.loc[condicio_candidat, 'candidat_Dona_rel'] = \
    df_candidatures_partits.loc[condicio_candidat, 'candidat_Dona'] / df_candidatures_partits.loc[condicio_candidat, 'candidat_Total']

df_candidatures_partits.loc[condicio_candidat, 'candidat_Home_rel'] = \
    df_candidatures_partits.loc[condicio_candidat, 'candidat_Home'] / df_candidatures_partits.loc[condicio_candidat, 'candidat_Total']


df_candidatures_partits.loc[condicio_electe, 'electe_Dona_rel'] = \
    df_candidatures_partits.loc[condicio_electe, 'electe_Dona'] / df_candidatures_partits.loc[condicio_electe, 'electe_Total']

df_candidatures_partits.loc[condicio_electe, 'electe_Home_rel'] = \
    df_candidatures_partits.loc[condicio_electe, 'electe_Home'] / df_candidatures_partits.loc[condicio_electe, 'electe_Total']
```

```python
# aquesta funció te d'entrada la columna
def procesar_columna(columna):


  # crear noms de les columnes
  col_home = columna + '_Home'
  col_dona = columna + '_Dona'
  col_total = columna + '_Total'
  col_home_rel = col_home + '_rel'
  col_dona_rel = col_dona + '_rel'

  columnes = [col_home, col_dona, col_total, col_home_rel, col_dona_rel]
  display(columnes)

  # condició per al calcul
  condicio = df_candidatures_partits[col_total] > 0

  # Tractar nuls
  df_candidatures_partits[col_home_rel] = np.nan
  df_candidatures_partits[col_dona_rel] = np.nan

  # calcular columnes
  df_candidatures_partits.loc[condicio, col_dona_rel] = \
    df_candidatures_partits.loc[condicio, col_dona] / df_candidatures_partits.loc[condicio, col_total]

  df_candidatures_partits.loc[condicio, col_home_rel] = \
    df_candidatures_partits.loc[condicio, col_home] / df_candidatures_partits.loc[condicio, col_total]


  display(df_candidatures_partits[columnes].describe())
  print(df_candidatures_partits[columnes].hist())
```

```python
# per cada columna calculs, invocar la funcio
for columna in calculs:
  procesar_columna(columna)
```
